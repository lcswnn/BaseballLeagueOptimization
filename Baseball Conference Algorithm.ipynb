```python
#Import data
import pandas as pd

baseballData = pd.read_csv('Stats.csv')
baseballData
```

```python
#initializing data
gamesPlayed = baseballData['G']
Wins = baseballData['W']
Losses = baseballData['L']
winsOverLosses = baseballData['W-L%']
gamesAbove = baseballData['G>.500']
divsWon = baseballData['Divs']
pnntsWon = baseballData['Pnnts']
WS = baseballData['WS']
Playoffs = baseballData['Playoffs']
nFormerHOF = baseballData['HOF#']
Runs = baseballData['R']
Hits = baseballData['H']
atBats = baseballData['AB']
homeRuns = baseballData['HR']
batAvg = baseballData['BA']
runsAllowed = baseballData['RA']
ERA = baseballData['ERA']
RunDif = baseballData['RD']
```

```python
#clustering algorithm
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

baseballData.dropna(inplace=True)
```

```python
#standardize data and add to table
scaler = StandardScaler()
baseballData[['G_T', 'W-L%_T', 'Runs_T', 'Hits_T', 'BA_T', 'RA_T', 'ERA_T', 'RD_T']] = scaler.fit_transform(baseballData[['G', 'W-L%', 'R', 'H', 'BA', 'RA', 'ERA', 'RD']])
baseballData
```

```python
#plot elbow point
def optimize_k_means(data, max_k):
    means= []
    inertias = []
    
    for k in range(1, max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        
        means.append(k)
        inertias.append(kmeans.inertia_)
        
    #Generate elbow
    fig = plt.subplots(figsize=(10,5))
    plt.plot(means, inertias, 'o-')
    plt.xlabel('Number of CLusters')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()
```

```python
optimize_k_means(baseballData[['G_T', 'W-L%_T', 'Runs_T', 'Hits_T', 'BA_T', 'RA_T', 'ERA_T']], 10)
```
